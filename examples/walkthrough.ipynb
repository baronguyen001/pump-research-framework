# Fictional Pumpscore Walkthrough

This notebook scores a fictional token with hand-entered research. It does not fetch data, call APIs, or make trade recommendations.

```python
from pumpscore import evaluate_checklist, fomo_decision, goldilocks_gate, score, suggest_size
from pumpscore.io import _read_mapping, load_scorecard

path = "examples/scorecard_example.yaml"
raw = _read_mapping(path)
card = load_scorecard(path)
score(card)
```

```python
evaluate_checklist(raw["patterns"], raw["red_flags"])
goldilocks_gate(card)
fomo_decision(card)
```

```python
suggest_size(card.total(), bankroll=10_000, asset_class="midcap")
```
